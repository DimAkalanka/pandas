```python
df.fillna(0) #Fill with zeros fro the place of missing values
df.fillna(df.mean()) #Fill with mean value
df.fillna(df.median()) #Fill with median value
df.fillna(df.mode()) #Fill with mode
df.fillna(df.min()) #Fill with min value
df.fillna(df.max()) #Fill with max value
df.fillna(df.std()) #Fill with std
df.fillna(df.var()) #Fill with var
df.fillna(df.sum()) #Fill with sum
df.fillna(df.count()) #Fill with count
df.fillna(df.describe()) #Fill with describe
df.fillna(df.corr()) #Fill with corr
df.fillna(df.cov()) #Fill with cov
df.fillna(df.skew()) #Fill with skew
df.fillna(df.kurt()) #Fill with kurt
df.fillna(df.quantile()) #Fill with quantile
df.fillna(df.nunique()) #Fill with nunique
df.fillna(df.value_counts()) #Fill with value_counts
df.fillna(df.apply(lambda x: x.value_counts().index[0])) #Fill with apply
df.fillna(df.interpolate()) #Fill with interpolate
df.fillna(df.ffill()) #Fill with ffill
df.fillna(df.bfill()) #Fill with bfill
df.fillna(df.interpolate(method='linear')) #Fill with interpolate
df.fillna(df.interpolate(method='quadratic')) #Fill with interpolate
df.fillna(df.interpolate(method='cubic')) #Fill with interpolate
df.fillna(df.interpolate(method='spline', order=2)) #Fill with interpolate
df.fillna(df.interpolate(method='spline', order=3)) #Fill with interpolate
df.fillna(df.interpolate(method='spline', order=4)) #Fill with
df.fillna(df.interpolate(method='spline', order=5)) #Fill with
```

# **Data Normalization**

**Normalization** is used in data preprocessing to scale the features of your data so that they have similar ranges. This process is crucial for various reasons, particularly when you're working with machine learning models. Here's why normalization is important:

1. **Improving Model Performance:**
Many machine learning algorithms, such as gradient descent-based models (e.g., linear regression, logistic regression, neural networks), perform better when the input features are on a similar scale. If the features are not normalized, features with larger ranges can dominate the cost function, making it harder for the model to converge.
2. **Ensuring Faster Convergence:**
Normalizing the data can help the model converge more quickly during training. This is because normalized data can make the optimization process smoother, preventing the algorithm from oscillating and thus speeding up convergence.
3. **Preventing Bias:**
Some machine learning algorithms, like k-Nearest Neighbors (k-NN) and Support Vector Machines (SVM), rely on distance measures (e.g., Euclidean distance). If the features have different scales, features with larger ranges will have more influence on the distance computation, potentially biasing the model. Normalization ensures that each feature contributes equally.
4. **Improving Interpretability:**
In some cases, normalized data can be more interpretable. For example, when comparing features directly or when visualizing data, having features on the same scale makes it easier to understand and compare them.
5. **Reducing Numerical Instability:**
Normalization can help avoid numerical issues in algorithms that are sensitive to large differences in the range of data values, like those that involve matrix operations (e.g., calculating the inverse of a matrix).
6. **Required by Certain Algorithms:**
Some algorithms, such as Principal Component Analysis (PCA), require data normalization as part of the process because they are sensitive to the variance in the data.

**When to Normalize:**

**Normalize when:**
The features have different ranges.
You are using algorithms that are sensitive to the scale of data (e.g., k-NN, SVM, neural networks).
Your model's training process involves gradient descent.

**Don't normalize when:**

The algorithm is scale-invariant (e.g., tree-based methods like decision trees, random forests).
The feature scales are already similar or have meaningful differences.

# **Binning**

**Categorization:**
Binning converts continuous variables into categorical ones, making it easier to analyze and interpret data. For instance, you might bin ages into ranges like 0-18, 19-35, 36-50, etc., to analyze patterns across age groups.

With use of Binning

   can use for noise reduction, improve model performance, facilitate visualizations


```python
bins = np.linspace(min(df['Price']), max(df['Price']), 4)
group_names = ['Low', 'Medium', 'High']
df['Price_binned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True)
```
