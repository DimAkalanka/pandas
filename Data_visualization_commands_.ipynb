# **Pseudo Color Plot**

A pseudocolor plot displays matrix data as an array of colored cells (known as faces). This plot is created as a flat surface in the x-y plane. The surface is defined by a grid of x and y coordinates that correspond to the corners (or vertices) of the faces. Matrix C specifies the colors at the vertices. The color of each face depends on the color of one of its four surrounding vertices. Of the four vertices, the one that comes first in the x-y grid determines the color of the face.

In this course, you use the pcolor plot for visualizing the contents of a pivot table that has been grouped on the basis of 2 parameters. Those parameters then represent the x and y-axis components that create the grid. The values in the pivot table are the average values of a third parameter. These values act as the code for the color the cell is going to take.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2D array (e.g., 5x5 matrix of random values)
data = np.random.rand(5, 5)

plt.pcolor(data, cmap='viridis')  # Use pcolor with 2D data
plt.colorbar()  # Add color bar
plt.show()
```

# **Residual plots**

What is a Residual Plot?
A residual plot is a graph that shows the residuals on the vertical axis and the independent variable on the horizontal axis. The residual is the difference between the actual data point (observed value) and the predicted value from the regression model.

Residuals help in understanding how well a regression model fits the data. Ideally, if the regression model fits the data well, the residuals should be randomly distributed around zero (no clear pattern).

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame (Height and Weight)
data = pd.DataFrame({
    'height': np.random.rand(50) * 50 + 150,  # Random heights (in cm)
    'weight': np.random.rand(50) * 30 + 50    # Random weights (in kg)    meaning of this is 50 random values in range of 50 to 80
})

# Create the residual plot
sns.residplot(x='height', y='weight', data=data)

plt.title("Residual Plot of Height vs. Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Residuals")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'header_1': [1, 2, 3, 4, 5],
    'header_2': [2, 4, 5, 4, 5]
})

# Prepare data
X = df[['header_1']]
y = df['header_2']

# Fit model
model = LinearRegression()
model.fit(X, y)

# Predict
df['predicted'] = model.predict(X)
df['residuals'] = df['header_2'] - df['predicted']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot residuals
sns.residplot(x=df['header_1'], y=df['header_2'])

plt.title("Residual Plot")
plt.xlabel("Header 1")
plt.ylabel("Residuals")
plt.show()
```

# **KDE plot**

A Kernel Density Estimate (KDE) plot is a graph that creates a probability distribution curve for the data based upon its likelihood of occurrence on a specific value. This is created for a single vector of information. It is used in the course in order to compare the likely curves of the actual data with that of the predicted data.

```python
sns.kdeplot(X)
```

# **Correlation**

Pearson correlation

```python
from scipy import stats

# Example DataFrame
import pandas as pd

df = pd.DataFrame({
    'horsepower': [100, 150, 200, 250, 300],
    'price': [200, 2000, 30000, 3500, 40000]
})

# Calculate Pearson correlation coefficient and p-value
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, "with a P-value of P =", p_value)
```

# **Chi Squad Test**

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Load the CSV file into a pandas DataFrame
# Replace 'your_file.csv' with the actual file path
data = pd.read_csv('your_file.csv')

# Separate features (X) and target (y)
# Replace 'target_column' with the actual name of your target column
X = data.drop(columns=['target_column'])  # Drop the target column to get the features
y = data['target_column']  # Target column

# Apply chi-squared test for feature selection
chi2_selector = SelectKBest(score_func=chi2, k=10)  # Select top 10 features
X_kbest = chi2_selector.fit_transform(X, y)

# Print the chi-squared scores of each feature
chi_scores = pd.Series(chi2_selector.scores_, index=X.columns)
print("Chi-squared scores:\n", chi_scores)

# Print the names of the selected features
selected_features = X.columns[chi2_selector.get_support()]
print("Selected features:\n", selected_features)

# Optionally, you can save the reduced dataset with selected features
X_selected = pd.DataFrame(X_kbest, columns=selected_features)
X_selected.to_csv('selected_features.csv', index=False)
```
